# Visualising your 
For more options see: https://github.com/aligfellow/xyzrender

```python
from ase.build import molecule
from xyzrender import load, render, render_gif

# Prepare or read your structure
atoms = molecule("C6H6")
atoms.write("benzene.xyz")

mol = load("benzene.xyz")
```

```python
render(mol)                          # displays inline in Jupyter
# render(mol, output="benzene.svg")   # Save as SVG/PNG/PDF
```

# Rotation GIF

```python
render_gif(mol, gif_rot="y")         
```
