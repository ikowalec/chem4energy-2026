# QM Input Example
This notebook shows how to write a Quantum ESPRESSO input file using ASE.

See the list of available calculators here: https://ase-lib.org/ase/calculators/calculators.html#module-ase.calculators

Read the atomic structure.

````python
from ase.io import read, write
atoms = read("data/TiNiO3.cif")
````

Define Quantum ESPRESSO input parameters.

````python
input_data = {
    "calculation": "relax",
    "restart_mode": "from_scratch",
    "tprnfor": True,
    "etot_conv_thr": 1e-5,
    "forc_conv_thr": 1e-4,
    "ecutwfc": 60,
    "ecutrho": 480,
    "input_dft": "rpbe",
    "vdw_corr": "dft-d3",
    "occupations": "smearing",
    "degauss": 0.01,
    "smearing": "cold",
    "conv_thr": 1e-8,
    "mixing_mode": "local-TF",
    "mixing_beta": 0.35,
    "diagonalization": "david",
    "ion_dynamics": "bfgs",
    "bfgs_ndim": 6,
    "startingwfc": "random",
}
````

Define pseudopotentials and write the input file.

````python
pseudopotentials = {
    "Ti": "Ti.pz-vbc.UPF",
    "Ni": "Ni.pz-vbc.UPF",
    "O": "O.pz-vbc.UPF",
}

write(
    "data/pw.in",
    atoms,
    input_data=input_data,
    pseudopotentials=pseudopotentials,
    format="espresso-in",
)
````

Reference: https://ase-lib.org/ase/calculators/espresso.html

Optional Hubbard example:

```python
additional_cards = ["HUBBARD (ortho-atomic)", "U Mn-3d 5.0", "U Ni-3d 6.0"]

write(
    "pw_hubbard.in",
    atoms,
    input_data=input_data,
    additional_cards=additional_cards,
    format="espresso-in",
)
```
